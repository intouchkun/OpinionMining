After data mining part is finish, We use opinion analysis that use with sentiwordnet. We read the file that collect tweets data.

```python
with open('OutputOld.txt') as f:
    lines = f.readlines()

inputs=lines[0]
wordOld=inputs.split(',')
wordOld=np.asarray(wordOld)
print len(wordOld)
print type(wordOld)
print wordOld[450]
```

```python
with open('OutputNew.txt') as f:
    lines = f.readlines()

wordNew=[]
inputs=lines[0]
wordNewsInput=inputs.split(',')

for wordIndex in wordNewsInput :
    wordIndex=re.sub(r'(^ +)',"",wordIndex)
    wordIndex=re.sub(r'(^[[])',"",wordIndex)
    wordIndex=re.sub(r'(^u)',"",wordIndex)
    wordIndex=re.sub(r'(^\'|")',"",wordIndex)
    wordIndex=re.sub(r'(\'|")$',"",wordIndex)
    wordNew.append(wordIndex)

wordNew=np.asarray(wordNew)
print wordNew[161]
print len(wordNew)
print type(wordNew)
```

This part is calculating score for each tweet by using Sentiwordnet library. we have to tokenize and pos_tag for getting Part Of Speech (POS) .

pScore/pword/ptweet = Positive Score / positive word / Number of positive tweet

nScore/nword/ntweet = Negative Score / Negative word / Number of Negative tweet

oScore/oword/otweet = Objective Score / Objective word / Number of Objective tweet

ssword = Senti_Synset word

```python
wordScore=[]
pWord=[]
nWord=[]
oWord=[]
pTweet=0
nTweet=0
oTweet=0
#wordsNew=wordNew[:50]
for words in wordNew:
    pScore=0
    nScore=0
    oScore=0
    token = nltk.word_tokenize(words)
    tagged = nltk.pos_tag(token)
    for wordToken,wordPOS in tagged:
        if(wordPOS in ["JJ","RB","NN","VB"]):
            if(wordPOS=="JJ"):
                wordSynset=wordToken+".a.01"
            elif(wordPOS=="RB"):
                wordSynset=wordToken+".r.01"
            elif(wordPOS=="NN"):
                wordSynset=wordToken+".n.01"
            else:
                wordSynset=wordToken+".v.01"
            try :
                ssword = swn.senti_synset(wordSynset)
                pScore+=ssword.pos_score()
                nScore+=ssword.neg_score()
                oScore+=ssword.obj_score()
            except :
                ssword=swn.senti_synsets(wordToken)
                if(len(ssword)>0):
                    pScore+=ssword[0].pos_score()
                    nScore+=ssword[0].neg_score()
                    oScore+=ssword[0].obj_score()
                    #print "pos"+str(wordSynset[0].pos_score())+"neg"+str(wordSynset[0].neg_score())+"obj"+str(wordSynset[0].obj_score())
        '''
        else :
            ssword=swn.senti_synsets(wordToken)
            if(len(ssword)>0):
                pScore+=ssword[0].pos_score()
                nScore+=ssword[0].neg_score()
                oScore+=ssword[0].obj_score()
                #print "pos"+str(wordSynset[0].pos_score())+"neg"+str(wordSynset[0].neg_score())+"obj"+str(wordSynset[0].obj_score())
        '''
    #wordScore.append(str(pScore)+","+str(nScore)+","+str(oScore))
    if(pScore<0.7 and nScore<0.7):
        oWord.append(words)
        oTweet+=1
    elif (pScore>nScore):
        pWord.append(words)
        pTweet+=1
    elif(nScore>pScore):
        nWord.append(words)
        nTweet+=1
    else:
        oWord.append(words)
        oTweet+=1
    
```
